```python
marks = [80, 0, 90, 60] # boolean_mask
credits = [4, 0, 2, 1]


weighed_avg = 0
sum_credit = 0
for m, c in zip(marks, credits):
    weighed_avg += m * c
    sum_credit += c


weighed_avg = weighed_avg/sum_credit
print(weighed_avg)
```

Follow up: Would IQR Change if instead of dropping missing value, we impute it with mean?

```python
import numpy as np

data = np.array([25, 28, 28, 30, 33, 36, 37, 37, 100, 38, 39, 41, 44, None, 47])
data
```

```python
# step 1: treat null value
data = data[data != None] # bool mask and filter
data
```

```python
# step2: IQR

# sort the value
data = np.sort(data)

# 25th percentile
q1 = np.percentile(data, 25)

# 75th percentile
q3 = np.percentile(data, 75)


print("IQR", q3 - q1)
```

```python
import statistics

data = [18, 22, 22, 25, 27, 28, 29, 30, 31, 95]
```

```python
mean = statistics.mean(data)
median = statistics.median(data)
mode = statistics.mode(data)
std = statistics.stdev(data)
range = max(data) - min(data)

# print everything
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", std)
print("Range:", range)
```
